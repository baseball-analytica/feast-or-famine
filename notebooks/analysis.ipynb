```python
import pandas as pd
import pybaseball

pybaseball.cache.enable()
```

```python
from typing import Any

data: dict[tuple[int, str], dict[str, Any]] = {}
```

```python
from tqdm import tqdm
import time

for year in tqdm(range(1998, 2026)):
    print(f"starting year {year}")
    all_teams = pybaseball.team_pitching(year)["Team"].unique().tolist()

    for team in tqdm(all_teams):
        print(f"starting team {team}")
        time.sleep(1)
        team_data = pybaseball.schedule_and_record(year, team)

        team_data["rdiff"] = team_data["R"] - team_data["RA"]
        fof = team_data["rdiff"].abs().mean()
        rs = team_data["R"].mean()
        ra = team_data["RA"].mean()
        rdiff = team_data["rdiff"].mean()
        wpct = (team_data.loc[(team_data["W/L"] == "W") | (team_data["W/L"] == "W-wo"), "W/L"].size) / team_data["W/L"].size

        data[(year, team)] = {
            "fof": fof,
            "rs": rs,
            "ra": ra,
            "rdiff": rdiff,
            "wpct": wpct,
        }

        print(f"finished team {team}")

    print(f"finished year {year}")

print("finished all years")
```

```python
print(data)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

mpl.rcParams["font.family"] = "monospace"

fig, ax = plt.subplots(figsize=(10, 6))

fig.suptitle("Feast-or-Famine Index vs. Runs Scored per Game", fontsize=16)

plt.title("All MLB team-seasons from 1998 through 2025; regular season only", pad=10)

ax.scatter(
    [d["fof"] for d in data.values()],
    [d["rs"] for d in data.values()],
    alpha=0.5,
    color="black"
)

ax.set_xlabel("FOF", fontsize=14)
ax.set_ylabel("RS/G", fontsize=14)

plt.yticks(
    np.linspace(2.5, 6.5, 17)
)

n = len(data)
trendline = np.polyfit([d["fof"] for d in data.values()], [d["rs"] for d in data.values()], 1)
trendline_fn = np.poly1d(trendline)
trendline_eq = f"y = {trendline[0]:.2f}x + {trendline[1]:.2f}"
r = np.corrcoef([d["fof"] for d in data.values()], [d["rs"] for d in data.values()])[0, 1]
r_squared = r**2

ax.plot(
    [d["fof"] for d in data.values()],
    trendline_fn([d["fof"] for d in data.values()]),
    color="red",
    label=trendline_eq,
    alpha=0.5
)

plt.grid(
    axis="both",
    alpha=0.5
)

plt.show()

print(f"n: {n}")
print(f"trendline: {trendline_eq}")
print(f"R: {r:.2f}, R^2: {r_squared:.2f}")
```
